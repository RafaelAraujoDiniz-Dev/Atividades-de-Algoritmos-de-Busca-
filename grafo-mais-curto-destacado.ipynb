```python
import networkx as nx
import matplotlib.pyplot as plt

# Cria o grafo direcionado
meu_grafo = nx.DiGraph()

# Dicionário de arestas do grafo
arestas_do_grafo = {
    'A': [('B', 2), ('C', 4)],
    'B': [('A', 2), ('C', 1), ('D', 7)],
    'C': [('A', 4), ('B', 1), ('E', 3)],
    'D': [('B', 7), ('E', 1)],
    'E': [('C', 3), ('D', 1)],
}

# Caminho mais curto de exemplo (igual ao da imagem)
caminho_em_destaque = ['A', 'B', 'C', 'E']

# Adiciona as arestas ao grafo com os respectivos pesos
for no_origem, lista_de_arestas in arestas_do_grafo.items():
    for no_destino, peso_aresta in lista_de_arestas:
        meu_grafo.add_edge(no_origem, no_destino, weight=peso_aresta)

# Fixe o layout para não mudar a cada execução
pos_nos = nx.spring_layout(meu_grafo, seed=42)

# Destaca as arestas do caminho mais curto
arestas_em_destaque = [(caminho_em_destaque[i], caminho_em_destaque[i+1]) for i in range(len(caminho_em_destaque)-1)]

# Desenha o grafo
plt.figure(figsize=(10, 6))
nx.draw(meu_grafo, pos_nos, with_labels=True, node_color='skyblue', node_size=2000, font_size=15, font_color='black')
nx.draw_networkx_edge_labels(meu_grafo, pos_nos, edge_labels={(u, v): d['weight'] for u, v, d in meu_grafo.edges(data=True)})
nx.draw_networkx_edges(meu_grafo, pos_nos, edgelist=arestas_em_destaque, edge_color='red', width=3)

# Título do gráfico
plt.title("Grafo com o Caminho Mais Curto Destacado (A → B → C → E)")
plt.show()
```
