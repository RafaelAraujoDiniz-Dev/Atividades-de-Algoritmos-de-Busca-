```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando uma árvore
estrutura_arvore = nx.DiGraph()

# Adicionando vértices e arestas
estrutura_arvore.add_edges_from([
    ('Raiz', 'N0'),
    ('Raiz', 'N1'),
    ('N0', 'N00'),
    ('N0', 'N01'),
    ('N1', 'N10'),
    ('N1', 'N11'),
])

# Definindo posições manuais (sem pygraphviz)
coordenadas_nos = {
    'Raiz': (0, 3),
    'N0': (-1.5, 2),
    'N1': (1.5, 2),
    'N00': (-2, 1),
    'N01': (-1, 1),
    'N10': (1, 1),
    'N11': (2, 1)
}

# Desenho
nx.draw(estrutura_arvore, coordenadas_nos, with_labels=True, node_size=2000, node_color='lightblue', arrows=True)
plt.title("Árvore Hierárquica (sem Graphviz)")
plt.axis('off')
plt.tight_layout()
plt.show()

# Função para obter propriedades do grafo (sem list comprehension)
def propriedades_grafo(grafo):
    # Verificar se o grafo é acíclico
    aciclico = True
    try:
        ciclos_encontrados = list(nx.find_cycle(grafo, orientation='original'))
        if ciclos_encontrados:
            aciclico = False
    except nx.exception.NetworkXNoCycle:
        aciclico = True

    print(f"O grafo é acíclico? {aciclico}")

    # Encontrar raiz (nós com grau de entrada 0)
    nos_raiz = []
    for vertice in grafo.nodes():
        if grafo.in_degree(vertice) == 0:
            nos_raiz.append(vertice)
    print(f"Raiz do grafo: {nos_raiz}")

    # Encontrar folhas (nós com grau de saída 0)
    nos_folha = []
    for vertice in grafo.nodes():
        if grafo.out_degree(vertice) == 0:
            nos_folha.append(vertice)
    print(f"Folhas do grafo: {nos_folha}")

    # Grau de entrada
    graus_entrada = {}
    for vertice in grafo.nodes():
        graus_entrada[vertice] = grafo.in_degree(vertice)

    # Grau de saída
    graus_saida = {}
    for vertice in grafo.nodes():
        graus_saida[vertice] = grafo.out_degree(vertice)

    print(f"Graus de entrada: {graus_entrada}")
    print(f"Graus de saída: {graus_saida}")

    return aciclico, nos_raiz, nos_folha, graus_entrada, graus_saida

# Chamada da função
propriedades = propriedades_grafo(estrutura_arvore)


# Supondo que grafo seja o grafo definido anteriormente

# Função para encontrar a raiz, folhas, nós de entrada e saída
# FACA AQUI A IMPLEMENTACAO DA FUNCAO SEM USAR LIST COMPREHENSION, OU SEJA, USE LACOS FOR
# CRIE SEUS PROPRIOS NOMES DAS VARIAVEIS
def propriedades_grafo(grafo):
    # Identificar se o grafo é acíclico
    aciclico = []
    print(f"O grafo é acíclico? {aciclico}")

    # Identificar a raiz (vértices com grau de entrada igual a zero)
    nos_raiz = []
    print(f"Raiz do grafo: {nos_raiz}")

    # Identificar as folhas (vértices com grau de saída igual a zero)
    nos_folha = []
    print(f"Folhas do grafo: {nos_folha}")

    # Calcular os graus de entrada e saída
    graus_entrada = {}
    graus_saida = {}
    print(f"Graus de entrada: {graus_entrada}")
    print(f"Graus de saída: {graus_saida}")

    return aciclico, nos_raiz, nos_folha, graus_entrada, graus_saida

propriedades = propriedades_grafo(estrutura_arvore)
```
