```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

# Lista de adjacência
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'E'],
    'D': ['B'],
    'E': ['C']
}

# Criar grafo não dirigido
G = nx.Graph()
for node in graph:
    for neighbor in graph[node]:
        G.add_edge(node, neighbor)

# Visualização do grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='lightblue', font_weight='bold')
plt.title("Grafo Não Dirigido")
plt.show()

# Função BFS com passo a passo
def bfs(graph, start):
    visited = set()
    fila = deque()
    ordem_visita = []

    print(f"Passo 1: Fila = [{start}], Visitados = {{}}")
    fila.append(start)

    while fila:
        atual = fila.popleft()
        if atual not in visited:
            visited.add(atual)
            ordem_visita.append(atual)
            print(f"\nDesenfileirar {atual}: Visitados = {visited}")
            for vizinho in graph[atual]:
                if vizinho not in visited:
                    fila.append(vizinho)
                    print(f"  Enfileira vizinho não visitado: {vizinho}")
            print(f"Fila atual = {list(fila)}")
    
    print(f"\nBusca finalizada. Ordem de visita: {' → '.join(ordem_visita)}")
    return ordem_visita

# Executar BFS
ordem = bfs(graph, 'A')
```
