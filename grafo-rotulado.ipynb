```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo direcionado
grafo_rotulado = nx.DiGraph()

# Adicionando vértices e arestas do grafo
grafo_rotulado.add_edge('0', '1', label='Φ')
grafo_rotulado.add_edge('1', '2', label='Γ')
grafo_rotulado.add_edge('0', '2', label='Ψ')

# Rotulação dos vértices
rotulos_vertices = {'0': 'φ', '1': 'γ', '2': 'ψ'}
for vertice, rotulo in rotulos_vertices.items():
    grafo_rotulado.nodes[vertice]['label'] = rotulo

# Desenhando o grafo
posicoes = nx.circular_layout(grafo_rotulado)  # Posicionamento circular para os nós
nx.draw(grafo_rotulado, posicoes, with_labels=False, node_size=2000, node_color='white', edgecolors='black')

# Desenhando os rótulos dos vértices
for vertice, (x, y) in posicoes.items():
    plt.text(x, y, f'{vertice}/{grafo_rotulado.nodes[vertice]["label"]}', fontsize=12, ha='center', va='center')

# Desenhando os rótulos das arestas
rotulos_arestas = nx.get_edge_attributes(grafo_rotulado, 'label')
nx.draw_networkx_edge_labels(grafo_rotulado, posicoes, edge_labels=rotulos_arestas, font_color='black')

# Mostrando o grafo
plt.axis('off')  # Desliga os eixos
plt.show()
```
