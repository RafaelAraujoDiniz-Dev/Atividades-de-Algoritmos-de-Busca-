```python
import networkx as nx
import matplotlib.pyplot as plt

# Defina o grafo como um dicionrio
grafo_direcionado = {
    'A': [('B', 2), ('C', 4)],
    'B': [('A', 2), ('C', 1), ('D', 7)],
    'C': [('A', 4), ('B', 1), ('E', 3)],
    'D': [('B', 7), ('E', 1)],
    'E': [('C', 3), ('D', 1)],
}

# Exemplo de caminho mais curto (pode ser gerado por Dijkstra)
caminho_mais_curto = ['A', 'C', 'E']

grafo_nx = nx.DiGraph()
for vertice, arestas in grafo_direcionado.items():
    for vizinho, peso in arestas:
        grafo_nx.add_edge(vertice, vizinho, weight=peso)

posicoes = nx.spring_layout(grafo_nx)
rotulos_arestas = {(u, v): d['weight'] for u, v, d in grafo_nx.edges(data=True)}

plt.figure(figsize=(10, 6))
nx.draw(grafo_nx, posicoes, with_labels=True, node_color='skyblue', node_size=2000, font_size=15, font_color='black')
nx.draw_networkx_edge_labels(grafo_nx, posicoes, edge_labels=rotulos_arestas, font_size=12)

if caminho_mais_curto:
    arestas_no_caminho = [(caminho_mais_curto[i], caminho_mais_curto[i + 1]) for i in range(len(caminho_mais_curto) - 1)]
    nx.draw_networkx_edges(grafo_nx, posicoes, edgelist=arestas_no_caminho, edge_color='red', width=3)

plt.title("Grafo com Caminho Mais Curto Destacado (A  E)")
plt.show()
```
